# Showing Plots

In Jupyter Notebooks, plots are automatically rendered and displayed inline when a plotting command (like plt.plot() or sns.scatterplot()) is executed within a cell.

```python
# YOUR CODE GOES HERE - your visualization must be shown as an output at the end of this cell 
#                      (do not store the image - it should be showon after this cell once you run this cell)

# Count room types
room_type_counts = listings_df['room_type'].value_counts()

# Plot pie chart
plt.figure(figsize=(6, 6))
room_type_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen'])
plt.title('Distribution of Room Types')
plt.ylabel('')  # Hide y-label
plt.show()

plt.show()
```
